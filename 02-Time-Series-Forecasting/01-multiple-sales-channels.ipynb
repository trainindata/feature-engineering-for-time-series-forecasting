## Air Quality

```python
filename = '../datasets/AirQualityUCI_ready.csv'

data = pd.read_csv(filename)

data.head()
```

```python
data['Date_Time'] = pd.to_datetime(data['Date_Time'])

data['Date'] = data['Date_Time'].dt.date
data['Time'] = data['Date_Time'].dt.time

data.head()
```

```python
# shorten time span to make simpler figures

data = data[data['Date_Time'] < '2005-01-01']
```

```python
plot_vars = ['T', 'RH']

data.groupby('Date')[plot_vars].mean().plot(figsize=(10, 4))
plt.ylabel('Value - Concentration')
plt.xlabel('')
plt.title('Daily Temperature and Humidity')
```

```python
data.groupby('Date_Time')[plot_vars].mean().plot(figsize=(10, 4))
plt.ylabel('Value - Concentration')
plt.xlabel('')
plt.title('Hourly Temperature and Humidity')
```
